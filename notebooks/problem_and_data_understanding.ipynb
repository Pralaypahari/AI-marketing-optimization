## Problem Statement

The goal of this project is to optimize marketing budget allocation across
multiple channels in order to maximize Gross Merchandise Value (GMV).

The challenge is that marketing channels have:
- Delayed effects (adstock)
- Diminishing returns (saturation)
- Interactions with seasonality and promotions

Before modeling, this notebook focuses on understanding the data and
business context.



## Business Context: Marketing Channels

Marketing channels represent different platforms or mediums where marketing
budget is allocated to acquire or engage customers.

In an e-commerce setting, typical channels include:
- Search advertising (high user intent)
- Social media advertising (discovery-driven)
- Display advertising (awareness-focused)
- Email marketing (owned, retention-focused)
- Influencer or affiliate marketing (trust-based reach)

Each channel differs in scale, cost structure, user intent, and response to
spend, which makes channel-level modeling essential for effective budget
allocation.


## Key Metrics: ROI and ROAS

Return on Investment (ROI) measures the efficiency of marketing spend by
comparing the value generated to the cost incurred.

Return on Ad Spend (ROAS) is a related metric that focuses specifically on the
sales generated per unit of advertising spend.

In this project, ROI-based interpretation is emphasized to support strategic
budget allocation decisions across channels.


## Why Channel-Level Modeling is Required

Different marketing channels influence customer behavior in different ways.
Some channels drive immediate high-intent conversions, while others contribute
to awareness or long-term demand.

Marketing Mix Modeling (MMM) explicitly models channels separately to account
for differences in delayed impact, diminishing returns, and overall contribution
to GMV. This enables more accurate and actionable budget optimization decisions.


## Time Series Behavior in E-commerce Sales

E-commerce sales data exhibits strong time-dependent patterns influenced by
long-term trends, seasonal effects, and short-term fluctuations.

Trend reflects the overall growth or decline in demand over time, while
seasonality captures recurring patterns such as weekly cycles, festivals, or
holiday periods. Random noise represents unexplained variation.

Marketing Mix Modeling (MMM) must account for these components to avoid
misattributing natural demand changes to marketing activities.


```python
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start = "2022-01-01", periods = 100, freq="W")
gmv = pd.Series(range(100))

plt.figure(figsize=(10, 4))
plt.plot(dates, gmv)
plt.title("gmv over time")
plt.xlabel("date")
plt.ylabel("gmv")
plt.show()
```

## Core Marketing Effects: Adstock and Saturation

Advertising effects are not instantaneous and do not scale linearly with spend.

Adstock represents the carryover effect of advertising, where marketing spend
continues to influence customer behavior and GMV over subsequent time periods.

Saturation captures the concept of diminishing returns, where increasing spend
beyond a certain point results in progressively smaller incremental gains in GMV.

Marketing Mix Modeling explicitly accounts for these effects to produce realistic
and actionable budget allocation insights.


## Exploratory Modeling with Random Forest

Random Forest regression is used in this project as an exploratory modeling
tool to identify important marketing channels and capture non-linear patterns
in the data.

The purpose of Random Forest modeling is to support intuition building and
validate channel relevance prior to applying Marketing Mix Modeling (MMM).
Random Forest outputs are not used directly for budget optimization or
causal decision-making.


## Model Explainability with SHAP

SHAP (SHapley Additive exPlanations) values are used to interpret the predictions
of the Random Forest model by quantifying the contribution of each marketing
channel to predicted GMV.

In this project, SHAP is applied to support exploratory analysis and channel
importance understanding. SHAP explanations are not treated as causal estimates
and are not directly used for budget optimization decisions.


## Bayesian Perspective and Uncertainty

Marketing data is inherently noisy and uncertain, making single-point estimates
insufficient for reliable decision-making.

Bayesian Marketing Mix Modeling incorporates prior knowledge and observed data to
produce posterior estimates with uncertainty intervals. These uncertainty
measures are critical for risk-aware marketing budget optimization and strategic
planning.
