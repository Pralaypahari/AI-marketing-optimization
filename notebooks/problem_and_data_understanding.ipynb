## Problem Statement

The goal of this project is to optimize marketing budget allocation across
multiple channels in order to maximize Gross Merchandise Value (GMV).

The challenge is that marketing channels have:
- Delayed effects (adstock)
- Diminishing returns (saturation)
- Interactions with seasonality and promotions

Before modeling, this notebook focuses on understanding the data and
business context.



## Business Context: Marketing Channels

Marketing channels represent different platforms or mediums where marketing
budget is allocated to acquire or engage customers.

In an e-commerce setting, typical channels include:
- Search advertising (high user intent)
- Social media advertising (discovery-driven)
- Display advertising (awareness-focused)
- Email marketing (owned, retention-focused)
- Influencer or affiliate marketing (trust-based reach)

Each channel differs in scale, cost structure, user intent, and response to
spend, which makes channel-level modeling essential for effective budget
allocation.


## Key Metrics: ROI and ROAS

Return on Investment (ROI) measures the efficiency of marketing spend by
comparing the value generated to the cost incurred.

Return on Ad Spend (ROAS) is a related metric that focuses specifically on the
sales generated per unit of advertising spend.

In this project, ROI-based interpretation is emphasized to support strategic
budget allocation decisions across channels.


## Why Channel-Level Modeling is Required

Different marketing channels influence customer behavior in different ways.
Some channels drive immediate high-intent conversions, while others contribute
to awareness or long-term demand.

Marketing Mix Modeling (MMM) explicitly models channels separately to account
for differences in delayed impact, diminishing returns, and overall contribution
to GMV. This enables more accurate and actionable budget optimization decisions.
